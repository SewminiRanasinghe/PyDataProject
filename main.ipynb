```python
import pandas as pd
import glob


csv_files = glob.glob("WorldCup_Stats/*.csv") 
dataframes = [pd.read_csv(file) for file in csv_files]


crick_df = pd.concat(dataframes, ignore_index=True)


crick_df.head()
```

Assume null recodes as which contains all columns are null

```python
#Data exploration
print(crick_df.info())
print(crick_df.describe())

# Remove duplicate records & Remove rows with null values
crick_df.drop_duplicates(inplace=True)
crick_df.dropna(how='all' ,inplace=True)

crick_df.head(10)
```

Since earlier assumption there are some null values in runs columns. Hence, it is unable to remove outlies.

```python
#Adding new columns to the DataFrame: match_status 
crick_df['match_status'] = crick_df['result'].apply(
    lambda x: 'abandoned' if 'abandoned' in x.lower() or 'no result' in x.lower() else 'played'
)

crick_df.head(10)
```

```python
#add new column as winning team 
def extract_winner(result):
    result = result.lower() 
    
    if 'won by' in result:
        return result.split(' won by ')[0].strip()  
    
    elif 'match tied' in result and 'won on boundary count' in result:
        
        team_name = result.split('match tied (')[1].split(' won on boundary count')[0]
        return team_name.strip() 
    
    return None  

crick_df['winning_team'] = crick_df['result'].apply(extract_winner)

crick_df.head(10)

```

To handle the case where a match is tied but a team wins on boundary count, in the winning team column shows the won country, and the case where a match is tied, in the winning team column shows the none and others shows won country 


```python
# Expanding 'best_bowlers' column
def split_best_bowlers(bowler_list):
    if pd.isna(bowler_list):
        return [None, None, None, None]
    bowlers = eval(bowler_list)  
    bowler_1, bowler_2 = (bowlers + [None, None])[:2]  
    bowler_1_name, bowler_1_wick = bowler_1.split(' - ') if bowler_1 else (None, None)
    bowler_2_name, bowler_2_wick = bowler_2.split(' - ') if bowler_2 else (None, None)
    return [bowler_1_name.strip(), bowler_1_wick.strip(), bowler_2_name.strip(), bowler_2_wick.strip()]

crick_df[['best_bowler_1', 'best_bowler_1_wick', 'best_bowler_2', 'best_bowler_2_wick']] = crick_df['best_bowlers'].apply(
    split_best_bowlers
).tolist()


crick_df.head(100-200)
```

```python
# Adding details of 'best_batters' columns
def split_best_batters(batter_list):
    if pd.isna(batter_list):
        return [None, None, None, None]
    batters = eval(batter_list)  # Converting string to list
    batter_1, batter_2 = (batters + [None, None])[:2]  # Handle cases with fewer than 2 entries
    batter_1_name, batter_1_runs = batter_1.split(' - ') if batter_1 else (None, None)
    batter_2_name, batter_2_runs = batter_2.split(' - ') if batter_2 else (None, None)
    return [batter_1_name.strip(), batter_1_runs.strip(), batter_2_name.strip(), batter_2_runs.strip()]

crick_df[['best_batter_1', 'best_batter_1_runs', 'best_batter_2', 'best_batter_2_runs']] = crick_df['best_batters'].apply(
    split_best_batters
).tolist()

crick_df.head(100-200)
```
